- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets,cluster

iris= datasets.load_iris()

X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
X[0:10]
```

```python
y
```

```python
from sklearn.cluster import KMeans
#n_clusters:cluster個數:想聚成幾類,n_init: 演算法預設會初始10個質心，實現演算法,random_state: 隨機生成簇中心的狀態條件。
estimators =  KMeans(n_clusters=3, n_init=10,random_state=0) 
test=estimators.fit_predict(X)
test
```

```python
import matplotlib.pyplot as plt
import numpy as np
f, ax = plt.subplots(figsize=(7.5, 7.5))
rgb = np.array(['r', 'g', 'b'])
# ax.scatter(X[:,0], X[:, 1], color=rgb[y]) #取 X[:,0], X[:, 1]分散不易群聚
ax.scatter(X[:, 2], X[:,3], color=rgb[test]) #取 X[:,2], X[:, 3]有群聚效果
ax.set_title("X")
```

```python
estimators.cluster_centers_ 
```

```python
f, ax = plt.subplots(figsize=(7.5, 7.5))
ax.scatter(X[:, 2], X[:,3], color=rgb[test])
ax.scatter(estimators.cluster_centers_[:, 2],estimators.cluster_centers_[:,3], marker='*', s=250,color='black', label='Centers')
ax.set_title("X")
ax.legend(loc='best')
```
