### 計算任意欄位的平均數及標準差

```python
df=df.iloc[0:10,[7,8,9]]
df
```

```python
listmn=df.mean()
listmn
```

### 平均數及標準差的直方圖

```python
#Mean:
Len1=len(df.mean())
lst1=[]
name1=[]
for i in range(0,Len1):
    X1=df.mean()[i]
    Y1=df.mean().index[i]
    name1.append(Y1)
    lst1.append(X1)
bar_with = 0.5
xlocations=np.array(range(len(lst1))) + bar_with
plt.title('Mean Value')
plt.bar(name1,lst1,color='lightblue',width=bar_with)
plt.show()

#Standart:
Len2=len(df.std())
lst2=[]
name2=[]
for i in range(0,Len2):
    X2=df.std()[i]
    Y2=df.std().index[i]
    name2.append(Y2)
    lst2.append(X2)
bar_with = 0.5
xlocations=np.array(range(len(lst2))) + bar_with
plt.title('Standart Value')
plt.bar(name2,lst2,color='lightgreen',width=bar_with)
plt.show()
```
